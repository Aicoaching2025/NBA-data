```python
from IPython import get_ipython
from IPython.display import display
# %%
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(suppress=True, precision = 2)
from google.colab import drive
# Mounting to /content/drive creates a 'MyDrive' folder.
drive.mount('/content/drive')

# Now you can access your file within 'MyDrive'
nba = pd.read_csv('/content/drive/MyDrive/nba_games.csv')
```

```python
# Subset Data to 2010 Season, 2014 Season
nba_2010 = nba[nba.year_id == 2010]
nba_2014 = nba[nba.year_id == 2014]

print(nba_2010.head())
print(nba_2014.head())
```

```python

#Create a series named knicks_pts_10 (fran_id = "Knicks")
knicks_pts_10 = nba_2010[nba_2010['fran_id'] == 'Knicks']['pts']
print(knicks_pts_10)

#Create another series names nets_pts_10(fran_id = "Nets")
nets_pts_10 = nba_2010[nba_2010['fran_id'] == 'Nets']['pts']
print(nets_pts_10)
```

```python
#Calculate the difference between the two teams’ average points scored and save the result as diff_means_2010.

nets_average = nets_pts_10.mean()
knicks_average = knicks_pts_10.mean()
diff_means_2010 = nets_average - knicks_average
print("The difference between the two teams’ average points scored:")
print(diff_means_2010)
print("    ")
print("The difference in mean is a measure of central tendency and describes the average separation between the values in two sets of data. It doesn't provide information about the nature or existence of a relationship between the variables fran_id and pts.")
```

```python
# Full Visualization Code for Comparing 2010 vs. 2014 Season

import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Replace with actual data
knicks_pts_10 = np.array([95, 102, 110, 87, 98, 120, 105, 115, 90, 102])  # Example data
nets_pts_10 = np.array([88, 94, 100, 85, 97, 112, 104, 108, 89, 101])  # Example data

knicks_pts_14 = np.array([98, 101, 104, 96, 99, 102, 100, 103, 97, 95])  # Example data
nets_pts_14 = np.array([100, 103, 105, 98, 101, 104, 102, 106, 99, 97])  # Example data

# Compute Averages
nets_average_10 = nets_pts_10.mean()
knicks_average_10 = knicks_pts_10.mean()
diff_means_2010 = nets_average_10 - knicks_average_10

nets_average_14 = nets_pts_14.mean()
knicks_average_14 = knicks_pts_14.mean()
diff_means_2014 = nets_average_14 - knicks_average_14

# Bar Chart Data
teams = ['Knicks (2010)', 'Nets (2010)', 'Knicks (2014)', 'Nets (2014)']
averages = [knicks_average_10, nets_average_10, knicks_average_14, nets_average_14]

# Create Bar Chart
plt.figure(figsize=(10, 6))
bars = plt.bar(teams, averages, color=['blue', 'lightgreen', 'darkblue', 'lightseagreen'], edgecolor='black')

# Adjust Y-Axis Scale for Better Visibility
plt.ylim(min(averages) - 5, max(averages) + 5)

# Add Labels & Title
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Average Points per Game', fontsize=12)
plt.title('Knicks vs. Nets: Average Points per Game (2010 vs 2014)', fontsize=14)

# Annotate Differences in Mean on the Bars
plt.text(0, knicks_average_10 + 1, f"{knicks_average_10:.2f} pts", ha='center', fontsize=12, fontweight='bold', color='white')
plt.text(1, nets_average_10 + 1, f"{nets_average_10:.2f} pts", ha='center', fontsize=12, fontweight='bold', color='white')
plt.text(2, knicks_average_14 + 1, f"{knicks_average_14:.2f} pts", ha='center', fontsize=12, fontweight='bold', color='white')
plt.text(3, nets_average_14 + 1, f"{nets_average_14:.2f} pts", ha='center', fontsize=12, fontweight='bold', color='white')

# Show the Plot
plt.show()
```

```python
# Determine if teams tend to win more games at home compared to away.
location_result_frequency = pd.crosstab(nba_2010.game_location, nba_2010.game_result)
print(location_result_frequency)
print("     ")
location_result_frequency_prop = pd.crosstab(nba_2010.game_location, nba_2010.game_result, normalize = True)
print(location_result_frequency_prop)
print("     ")
chi2, pval, dof, expected = chi2_contingency(location_result_frequency)
print(expected)
print(location_result_frequency_prop)
print("     ")

#calculate the covariance between forecast (538’s projected win probability) and point_diff
# Covariance is the number that is represented twice on in the matrix
point_diff_forecast_cov = np.cov(nba_2010.forecast, nba_2010.point_diff)
print(point_diff_forecast_cov)
print("The covariance is " + str(1.37) + ".")

# Calculate the correlation betwen two columns; forecast and point_diff
point_diff_forecast_corr = pearsonr(nba_2010.forecast, nba_2010.point_diff)
print("The correlation between the two columns; forecast and point_diff is ")
print(point_diff_forecast_corr)

# Create a scatter plot of forecast (on the x-axis) and point_diff (on the y-axis).
plt.clf()
scatter_forecast = plt.scatter(nba_2010.forecast, nba_2010.point_diff)
plt.xlabel('Forecasted Win Prob')
plt.ylabel('Point Differential')
plt.show()

print("The scatter plot does show a slight positive correlation which matches the corelation coefficient of  0.44 ")
```
