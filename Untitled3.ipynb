```python
# Additional Visualizations for NBA Data Analysis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Assuming your file is in 'MyDrive', adjust the path accordingly
file_path = '/content/drive/MyDrive/Codecademy_NBA_Trends_Project/nba_games.csv'  # Update with the correct path


np.set_printoptions(suppress=True, precision = 2)

nba = pd.read_csv(file_path)

# Subset Data to 2010 Season, 2014 Season
nba_2010 = nba[nba.year_id == 2010]
nba_2014 = nba[nba.year_id == 2014]

print(nba_2010.head())
print(nba_2014.head())

#Create a series named knicks_pts_10 (fran_id = "Knicks")
knicks_pts_10 = nba_2010[nba_2010['fran_id'] == 'Knicks']['pts']
print(knicks_pts_10)

#Create another series names nets_pts_10(fran_id = "Nets")
nets_pts_10 = nba_2010[nba_2010['fran_id'] == 'Nets']['pts']
print(nets_pts_10)


#Calculate the difference between the two teamsâ€™ average points scored and save the result as diff_means_2010.

nets_average = nets_pts_10.mean()
knicks_average = knicks_pts_10.mean()
diff_means_2010 = nets_average - knicks_average
print("The difference between the two teamsâ€™ average points scored:")
print(diff_means_2010)
print("    ")
print("The difference in mean is a measure of central tendency and describes the average separation between the values in two sets of data. It doesn't provide information about the nature or existence of a relationship between the variables fran_id and pts.")


# Create a set of overlapping histograms that can be used to compare the points scored for the Knicks compared to the Nets.

plt.hist(knicks_pts_10, alpha = 0.8, density = True, label = 'knicks', bins=30, color='blue', edgecolor='black')
plt.hist(nets_pts_10, alpha = 0.8, density = True, label = 'nets', bins=30, color='lightgreen', edgecolor='black')
plt.legend()
plt.title('2010 Season')
plt.show()
plt.clf()


#Data Analysis for the nets
#Create a series named knicks_pts_10 (fran_id = "Knicks")
knicks_pts_14 = nba_2014[nba_2014['fran_id'] == 'Knicks']['pts']
print(knicks_pts_14)

#Create another series names nets_pts_10(fran_id = "Nets")
nets_pts_14 = nba_2014[nba_2014['fran_id'] == 'Nets']['pts']
print(nets_pts_14)


#Calculate the difference between the two teamsâ€™ average points scored and save the result as diff_means_2010.

nets_average_14 = nets_pts_14.mean()
knicks_average_14 = knicks_pts_14.mean()
diff_means_2014 = nets_average_14 - knicks_average_14
print("The difference between the two teamsâ€™ average points scored during games in 2014:")
print(diff_means_2014)
print("The difference in mean got smaller.")

# Create a set of overlapping histograms that can be used to compare the points scored for the Knicks compared to the Nets.

plt.hist(knicks_pts_10, alpha = 0.8, density = True, label = 'knicks', bins=30, color='blue', edgecolor='black') # Changed normed to density
plt.hist(nets_pts_10, alpha = 0.8, density = True, label = 'nets', bins=30, color='lightgreen', edgecolor='black') # Changed normed to density
plt.legend()
plt.title('2010 Season')
plt.show()
plt.clf()


# ... (rest of your code) ...


# Create a set of overlapping histograms that can be used to compare the points scored for the Knicks compared to the Nets.

plt.hist(knicks_pts_14, alpha = 0.8, density = True, label = 'knicks', bins=30, color='orange', edgecolor='black') # Changed normed to density
plt.hist(nets_pts_14, alpha = 0.8, density = True, label = 'nets', bins=30, color='lightgreen', edgecolor='black') # Changed normed to density
plt.legend()
plt.title('2014 Season')
plt.show()
plt.clf()

# ... (rest of your code) ...

#For the remainder of this project, Iâ€™ll focus on data from 2010. I'll include all teams in the dataset and investigate the relationship between franchise and points scored per game.

# Options to investigate data include, box plots, histograms, contingency and frequency tables

sns.boxplot(data = nba, x = 'fran_id', y = 'pts')
plt.show()
plt.clf()
print("Correlation is a statistical measure that quantifies the degree to which two variables change together. The graph shows that the cental tendency of all the the variables are similar and I can loosely infer that the points scored is similar across teams.")
print("     ")
print("The Nets and the Knicks have a similar mean points cored for the season while the Spurs, Thunder and Celtics all have bery close values of the average points scored during the season.")
print("     ")

# Determine if teams tend to win more games at home compared to away.
location_result_frequency = pd.crosstab(nba_2010.game_location, nba_2010.game_result)
print(location_result_frequency)
print("     ")
location_result_frequency_prop = pd.crosstab(nba_2010.game_location, nba_2010.game_result, normalize = True)
print(location_result_frequency_prop)
print("     ")
chi2, pval, dof, expected = chi2_contingency(location_result_frequency)
print(expected)
print(location_result_frequency_prop)
print("     ")

#calculate the covariance between forecast (538â€™s projected win probability) and point_diff
# Covariance is the number that is represented twice on in the matrix
point_diff_forecast_cov = np.cov(nba_2010.forecast, nba_2010.point_diff)
print(point_diff_forecast_cov)
print("The covariance is " + str(1.37) + ".")

# Calculate the correlation betwen two columns; forecast and point_diff
point_diff_forecast_corr = pearsonr(nba_2010.forecast, nba_2010.point_diff)
print("The correlation between the two columns; forecast and point_diff is ")
print(point_diff_forecast_corr)

# Create a scatter plot of forecast (on the x-axis) and point_diff (on the y-axis).
plt.clf()
scatter_forecast = plt.scatter(nba_2010.forecast, nba_2010.point_diff)
plt.xlabel('Forecasted Win Prob')
plt.ylabel('Point Differential')
plt.show()

print("The scatter plot does show a slight positive correlation which matches the corelation coefficient of  0.44 ")

```
